## Ensemble methods to increase accuracy

```python
from sklearn.tree import DecisionTreeClassifier

DT= DecisionTreeClassifier(random_state=3)
DT.fit(X_train,y_train)

# Predict on validation data 
DT_predictions=DT.predict(X_test)
validation_accuracy=accuracy_score(y_test,DT_predictions)
print("Validation Accuracy of the model :-",validation_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier 

RF=RandomForestClassifier(random_state=21)
RF.fit(X_train,y_train)

# Predict on validation data 
RF_predictions=RF.predict(X_test)
validation_accuracy=accuracy_score(y_test,RF_predictions)
print("Validation Accuracy of the model :-",validation_accuracy)
```

```python
from sklearn.ensemble import BaggingClassifier

Bag=BaggingClassifier(random_state=89)
Bag.fit(X_train,y_train)

# Predict on validation data 
Bag_predictions=Bag.predict(X_test)
validation_accuracy=accuracy_score(y_test,Bag_predictions)
print("Validation Accuracy of the model :-",validation_accuracy)
```

```python
from sklearn.ensemble import AdaBoostClassifier

Boast=AdaBoostClassifier(random_state=79)
Boast.fit(X_train,y_train)

# Predict on validation data 
Boast_predictions=Boast.predict(X_test)
validation_accuracy=accuracy_score(y_test,Boast_predictions)
print("Validation Accuracy of the model :-",validation_accuracy)
```
